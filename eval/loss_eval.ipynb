```python
from evaluate_log import extract_metrics

df = extract_metrics('logs/training_log.log')

import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with the following columns:
# 'epoch', 'step', 'train_geh', 'train_mae', 'valid_geh', 'valid_mae'

# Plot GEH for training and validation
plt.figure(figsize=(10, 5))

# Define custom colors
train_color = (31/255, 119/255, 180/255)  # Blue for training
test_color = (214/255, 39/255, 40/255)    # Red for testing

# GEH Plot
plt.subplot(1, 2, 1)
plt.plot(df['epoch'] + df['step']/(max(df['step'])+1), df['train_geh'], label='Train MGEH', color=train_color)
plt.plot(df['epoch'] + df['step']/(max(df['step'])+1), df['valid_geh'], label='Test MGEH', color=test_color)
plt.title('MGEH')
plt.xlabel('Number of Training Epochs')
plt.ylabel('MGEH')
plt.legend()

# MAE Plot
plt.subplot(1, 2, 2)
plt.plot(df['epoch'] + df['step']/(max(df['step'])+1), df['train_mae'], label='Train MAE', color=train_color)
plt.plot(df['epoch'] + df['step']/(max(df['step'])+1), df['valid_mae'], label='Test MAE', color=test_color)
plt.title('MAE')
plt.xlabel('Number of Training Epochs')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
plt.savefig('learning_curve.png')
plt.show()
```
