```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "edge_features.csv"
data = pd.read_csv(file_path)

data = data[["Distance (km)", "Duration (s)", "Straight_line_distance", "Average_speed (km/h)", "Detour_factor"]]
colnames = ["Distance (km)", "Duration (s)", "Straight-line Distance (km)", "Average Speed (km/h)", "Detour Factor"]

# Number of columns
num_columns = data.shape[1]

# Create subplots
fig, axes = plt.subplots(1, num_columns, figsize=(15, 3))

# Plot each column as a histogram
for i, column in enumerate(data.columns):
    axes[i].hist(data[column].dropna(), bins=50, color='#1F77B4', alpha=1)  # Updated color
    axes[i].set_title(colnames[i])
    # axes[i].set_xlabel("Values")
    axes[i].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()
plt.savefig('edge_features.png')
plt.show()
```
